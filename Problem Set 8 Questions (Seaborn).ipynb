1. (5 pts) Use a Violin Plot to compare the distribution between 'survived' and 'pclass'. Use another Violin Plot to compare 'survived' and 'adult_male'. 'Survived' should be on the y-axis in both plots for easier visualization. What inferences can you make out of it? 

```python
sns.violinplot(data=titanic, y='survived', x='pclass')
plt.title('Survival on Titanic by Class')
plt.xlabel('Class')
plt.ylabel('Survival (1=Survived, 0=Died)');
```

```python
sns.violinplot(data=titanic, y='survived', x='adult_male')
plt.title('Survival on Titanic for Adult Males')
plt.xlabel('Adult Male')
plt.ylabel('Survival (1=Survived, 0=Died)');
```

- Adult males were most likely to not survive.
- Upper class passengers had higher survival rates than lower class passengers.

2. (5 pts) Use a Joint Plot to show the distribution between 'fare' and 'age'. Use survived as hue. 1. (7 pts) Select 'sepal_length','sepal_width','petal_length','petal_width' variables to plot a Pair Plot. It will be super helpful to make 'species' as your hue.

```python
pairplot_data = iris[['sepal_length','sepal_width','petal_length','petal_width', 'species']]
sns.pairplot(pairplot_data, hue='species')
```

2. (6 pts) What insights can you gather from these visualizations? As you can probably see, different clusters are formed by the variables, how do the clusters formed by the sepal length and width differ from those formed by the petal length and width?

- Petal length and width have a tight positive linear correlation while the correlation between sepal lenght and width is more diffuse.

3. (7 pts) Use the same variables from above and create a 'PairGrid' for the 'iris' dataset. On the diagonal maps, plot a histplot, on the upper maps plot a scatterplot, and on the lower maps plot a kdeplot. Also use 'species' as your hue. 

```python
import matplotlib.pyplot as plt
g = sns.PairGrid(pairplot_data, hue='species')

g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)

g.add_legend();
```
